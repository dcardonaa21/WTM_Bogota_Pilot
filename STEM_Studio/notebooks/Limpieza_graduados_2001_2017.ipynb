```python
graduados_2001_2017_dict = pd.read_excel("https://storage.googleapis.com/piloto_medellin/excel/GRADUADOS-2001-2017.xlsb", sheet_name=None, engine="pyxlsb")
```

```python
graduados_2001_2017_dict.keys()
```

```python
graduados_2001_2017_df = graduados_2001_2017_dict['Graduados 2001 a 2017']
graduados_2001_2017_df.head(11)
```

```python
graduados_2001_2017_df.drop(graduados_2001_2017_df.index[0:10], inplace=True)

```

```python
graduados_2001_2017_df = graduados_2001_2017_df.loc[graduados_2001_2017_df["Unnamed: 11"] == "MEDELLIN", :]
```

```python
filtered_graduados_2001_2017_df = graduados_2001_2017_df.filter(["Unnamed: 2", "Unnamed: 17", "Unnamed: 23", "Unnamed: 30", "Unnamed: 32"])
```

```python
rename_dict = {
    "Unnamed: 2":"INSTITUCION DE EDUCACION SUPERIOR (IES)", 
    "Unnamed: 17":"NIVEL DE FORMACION", 
    "Unnamed: 23":"NUCLEO BASICO DEL CONOCIMIENTO (NBC)", 
    "Unnamed: 30":"AÑO", 
    "Unnamed: 32":"GRADUADOS"}

```

```python
filtered_graduados_2001_2017_df.rename(columns=rename_dict, inplace=True)
filtered_graduados_2001_2017_df.head(10)
```

```python
filtered_graduados_2001_2017_df.index = list(range(0, len(filtered_graduados_2001_2017_df.iloc[:, 0])))
```
