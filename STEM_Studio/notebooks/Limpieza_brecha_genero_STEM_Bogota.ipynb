**2.1. ¿Cuál es la cantidad de hombres graduados en carreras STEM por año?**

```
hombres = df_estudiantes_bogota_2001_2018.loc[df_estudiantes_bogota_2001_2018["SEXO"]== "HOMBRE", :]
hombres.rename(columns={"GRADUADOS":"HOMBRES"}, inplace=True)
total_hombres_anual_df = pd.DataFrame(hombres.groupby(["AÑO"])["HOMBRES"].sum())
total_hombres_anual_df
```

**2.2. ¿Cuál es la cantidad de mujeres graduadas en carreras STEM por año?**

```
mujeres = df_estudiantes_bogota_2001_2018.loc[df_estudiantes_bogota_2001_2018["SEXO"]== "MUJER", :]
mujeres.rename(columns={"GRADUADOS":"MUJERES"}, inplace=True)
total_mujeres_anual_df = pd.DataFrame(mujeres.groupby(["AÑO"])["MUJERES"].sum())
total_mujeres_anual_df
```

**2.3. ¿Cuál es la distribución de la cantidad de graduados en STEM por año?**

```
total_graduados_anual_df = pd.DataFrame(df_estudiantes_bogota_2001_2018.groupby(["AÑO"])["GRADUADOS"].sum())
total_graduados_anual_df
```

**2.4. ¿Cuál es la brecha de graduados por año en carreras STEM?**

```
total_graduados_anual_df["DIFERENCIA"] = total_hombres_anual_df["HOMBRES"] - total_mujeres_anual_df["MUJERES"]
total_graduados_anual_df["BRECHA TOTAL STEAM"] = ((total_graduados_anual_df["DIFERENCIA"]/total_graduados_anual_df["GRADUADOS"])*100).round(0)
total_graduados_anual_df["AÑO"] = total_graduados_anual_df.index
total_graduados_anual_df.index = list(range(1, 18))
```

```
total_graduados_anual_df
```
